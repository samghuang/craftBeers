```python
beers = pd.read_csv('beers.csv')
beers
```

```python
breweries = pd.read_csv('breweries.csv')
breweries = breweries.rename(columns={"Unnamed: 0": "brewery_id"})
breweries.head()
```

```python
# merge beers and breweries 
beers_breweries = beers.merge(breweries, left_on='brewery_id', right_on='brewery_id')
beers_breweries = beers_breweries.drop(columns=['Unnamed: 0'])
beers_breweries = beers_breweries.rename(columns={"name_x": "beer", "name_y":"brewery"})
beers_breweries.head()
```

```python
beers_breweries.head()
```

```python
# How many beers are in the each state? 
state_count = beers_breweries[['state', 'city']].groupby('state').count()
```

```python
# What states have the most beers 
state_count.sort_values(by = 'city', ascending=False).head()
```

```python
# Top three states and the cities in from there 

# CO - Colorado
Colorado_beers = beers_breweries[beers_breweries.state == ' CO']
Colorado_cities = Colorado_beers.groupby('city').count()[['state']]

# OR - Oregon
Oregon_beers = beers_breweries[beers_breweries.state == ' OR']
Oregon_cities = Oregon_beers.groupby('city').count()[['state']]

# MI
Michigan_beers = beers_breweries[beers_breweries.state == ' MI']
Michigan_cities = Michigan_beers.groupby('city').count()[['state']]
```

```python
# Plot of the beers based on cities 
Colorado_cities.sort_values(by = 'state', ascending=False).head()

# Type of Beer enjoy of top states
Colorado_beers.groupby('style').count()[['beer']].head()
```

```python
Oregon_cities.sort_values(by = 'state', ascending=False).head()

# Type of Beer enjoy of top states
Oregon_beers.groupby('style').count()[['beer']].head()
```

```python
Michigan_cities.sort_values(by = 'state', ascending=False).head()

# Type of Beer enjoy of top states
Michigan_beers.groupby('style').count()[['beer']].head()
```
